seaborn is a data visualization library


```python
# importing all the libraries we learn so far
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seaborn has some built-in-dataset we can use for our practice like,
tips=sns.load_dataset('tips')
```

tips: A classic dataset containing information about tips given at a restaurant. It's often used to show relationships between variables like total_bill, tip, day, and time.

titanic: A historical dataset with demographic and survival information for passengers on the Titanic. It's a great example for demonstrating classification problems and visualizing categorical data.

iris: A famous dataset in machine learning and statistics, featuring measurements of sepals and petals for three species of iris flowers. It's ideal for illustrating clustering and classification techniques.

flights: This dataset contains the number of airline passengers per month from 1949 to 1960. It's excellent for creating time-series plots and heatmaps.

penguins: A dataset with measurements of different penguin species. It's a modern alternative to the iris dataset and is frequently used for scatter plots and statistical visualizations.

```python
# setting a theme for the plot (it is optional but sometimes helpful!)
sns.set_theme()
```

```python
#tips data vidualization
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='time',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',col='time',style='time',size='size')
```

```python
# Now we are importing a different dataset form sklearn dataset
from sklearn.datasets import load_diabetes
diabetes_bunch=load_diabetes()
diabetes = pd.DataFrame(data=diabetes_bunch.data, columns=diabetes_bunch.feature_names)
```

```python
diabetes.head()
```

```python
print(diabetes)
```

```python
sns.distplot(diabetes['sex'])
```

```python
# correlation
corr=diabetes.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True,cmap='RdYlGn')
```
