Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
# importing matPlotLib Library
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
print(x)
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
# plotting the data
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# adding "title" "x-labels" "y-labels"
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("sine wave")
```

```python
# Plotting a Parabola
x=np.linspace(-5,5,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
# changing the color of the plot
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'g')
plt.show()
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g.')
```

A bar plot, or bar chart, is a graph that uses rectangular bars to compare different categories. The length or height of each bar is proportional to the value it represents.

```python
# Drawing a Bar Plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,30,15]
ax.bar(langs,students)
plt.show()
```

The list [0,0,1,1] is a list of four numbers that specifies the position and size of the axes within the figure, as a fraction of the figure's width and height.

The first two numbers, 0, 0, define the bottom-left corner of the axes.

The last two numbers, 1, 1, define the width and height of the axes.

So, [0,0,1,1] means the axes will start at the bottom-left corner (0,0) of the figure and take up the entire width (1) and height (1) of the figure.

```python
# Drawing a pie chart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,30,15]
ax.pie(students,labels=langs,autopct='%1.2f%%')
plt.show()
```

autopct='%1.2f%%': This automatically calculates and displays the percentage for each slice directly on the chart. The string '%1.2f%%' formats the percentage to show two decimal places. The %% at the end is used to print a literal percent sign.

```python
# Drawing a scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker='*')
plt.show()
```

```python
plt.scatter(x,y,marker='*',color='g')
plt.show()
```

```python
# 3D scatter plot
fig=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='cool_r')
plt.show()
```
